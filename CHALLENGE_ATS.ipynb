```python
#CHALLENGE
# PROGRAM PEMANTAUAN ENERGI & PENGISIAN BATERAI PANEL SURYA

# energi hasil konversi dari panel surya
def hitung_energi(input_daya, efisiensi):
    """
    Menghitung energi keluaran (Wh) berdasarkan input daya dan efisiensi sistem.
    """
    energi_output = input_daya * efisiensi
    return energi_output

# Kondisi kapasitas batrai
def status_baterai(kapasitas):
    if kapasitas < 30:
        return "RENDAH "
    elif kapasitas < 80:
        return "NORMAL "
    elif kapasitas < 100:
        return "TINGGI "
    else:
        return "PENUH "


print(" Simulasi Pengisian Batrai Oleh Panel Surya ")

daya_masuk = float(input("Masukkan daya input dari panel surya (Wh): "))
efisiensi = float(input("Masukkan efisiensi sistem (0-1): "))

energi_harian = hitung_energi(daya_masuk, efisiensi)
print(f"\nEnergi yang dihasilkan per siklus: {energi_harian:.2f} Wh\n")

kapasitas_baterai = 0

# simulasi pengisian baterai
while kapasitas_baterai < 100:
    kapasitas_baterai += energi_harian / 10
    if kapasitas_baterai > 100:
        kapasitas_baterai = 100

    print(f"Kapasitas baterai saat ini: {kapasitas_baterai:.2f}% - {status_baterai(kapasitas_baterai)}")

    # notifikasi pengisian batrai
    if kapasitas_baterai >= 90 and kapasitas_baterai < 100:
        print(" WARNING: Baterai hampir penuh!")
    elif kapasitas_baterai == 100:
        print(" Baterai penuh. Pengisian berhenti.")
        break


print("- CATATAN DAYA LISTRIK HARIAN -")

# estimasi energi selama seminggu
for hari in range(1, 8):
    total_energi = energi_harian * hari
    print(f"Hari ke-{hari}: {total_energi:.2f} Wh terkumpul")


```
