```python
#2. Fungsi / Functions
# Program untuk menghitung efesiensi turbin

def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

#input 500 dan output 400
hasil = hitung_efisiensi(500, 400)
print("Efisiensi turbin adalah:", hasil * 100, "%")
```

```python


Penjelasannya:
Program ini dibuat untuk menghitung seberapa besar tingkat efisiensi sebuah turbin berdasarkan input energi dan output energi. Masalah yang diselesaikan oleh program ini adalah seberapa efektif turbin mengubah energi input menjadi energi output, yang penting dalam bidang energi dan ketenagalistrikan untuk menilai performa sistem pembangkit. Cara kerja program ini dimulai dengan kode hitung_efisiensi(input_energi, output_energi) yang berfungsi menghitung rasio antara energi keluaran terhadap energi masukan. Nilai ini kemudian dikembalikan oleh fungsi menggunakan perintah return efisiensi. Setelah itu, program memanggil fungsi tersebut dengan data contoh: (input energi = 500) dan (output energi = 400). Hasil perhitungan kemudian disimpan dalam variabel hasil, dan program menampilkan nilai efisiensi dalam bentuk persentase Bagian kode yang paling penting dalam program ini adalah baris efisiensi = output_energi / input_energi, karena di sinilah inti perhitungan efisiensi dilakukan. Tanpa baris ini, program tidak akan mampu menghitung seberapa besar energi output dibandingkan energi input. Secara keseluruhan, program ini sederhana namun efektif dalam membantu menganalisis performa turbin dengan cepat dan akurat.
```

```python
#4. Logika if Majemuk/bertingkat

tekanan_uap = 160.0          # contoh nilai tekanan
status_manual_override = False  # contoh status override

if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")
else:
    print("Katup Tertutup")
```

```python


Penjelasannya :
Program ini dibuat untuk mengontrol pembukaan katup pengaman (safety valve) berdasarkan dua kondisi utama. Katup hanya boleh terbuka jika tekanan uap melebihi 150 psi dan status manual override dalam keadaan False (tidak aktif). Dalam program Python, kedua kondisi tersebut digabung menggunakan operator logika and karena keduanya harus terpenuhi secara bersamaan agar katup dapat terbuka. Jika salah satu kondisi tidak terpenuhi, maka katup harus tetap tertutup untuk menjaga keamanan sistem.
Pada kode program, variabel tekanan_uap digunakan untuk menyimpan nilai tekanan aktual dalam satuan psi, sedangkan variabel status_manual_override menyimpan nilai boolean yang menunjukkan apakah sistem sedang dioperasikan secara manual atau tidak. Pernyataan if tekanan_uap > 150 and status_manual_override == False: digunakan untuk memeriksa kedua kondisi tersebut. Jika hasil evaluasi logika bernilai True, maka program akan mencetak pesan “Katup Terbuka” yang berarti sistem memenuhi syarat keamanan untuk membuka katup. Namun, jika salah satu atau kedua kondisi tidak terpenuhi (misalnya tekanan masih di bawah 150 psi atau mode manual override aktif), maka program akan menjalankan blok else dan menampilkan pesan “Katup Tertutup”.
Melalui logika ini, sistem pengaman berfungsi secara otomatis untuk mencegah terjadinya kesalahan operasional. Penggunaan struktur if majemuk dengan operator logika and membuat keputusan program lebih efisien karena hanya membuka katup pada kondisi yang benar-benar aman dan sesuai dengan kriteria yang telah ditentukan.

```

```python
#8. Fungsi / Modularisasi
def hitung_efisiensi_termal():
    """
    Fungsi untuk menghitung efisiensi termal turbin gas.
    Rumus: Efisiensi (%) = (energi_keluar / energi_masuk) * 100
    """
    print("="*45)
    print("    PROGRAM PENGHITUNG EFISIENSI TERMAL TURBIN GAS ")
    print("="*45)

    try:
        energi_masuk = float(input("Masukkan energi masuk (Joule): "))
        energi_keluar = float(input("Masukkan energi keluar (Joule): "))

        if energi_masuk <= 0:
            print(" Energi masuk tidak boleh nol atau negatif!")
            return

        efisiensi = (energi_keluar / energi_masuk) * 100

        print("\n HASIL PERHITUNGAN:")
        print(f"Efisiensi Termal = {efisiensi:.2f}%")

        # Analisis otomatis berdasarkan hasil
        if efisiensi >= 70:
            print(" Sangat efisien! Sistem bekerja optimal.")
        elif 40 <= efisiensi < 70:
            print(" Cukup efisien, tetapi masih bisa ditingkatkan.")
        else:
            print(" Efisiensi rendah, periksa kemungkinan kebocoran energi.")
        print("="*45)

    except ValueError:
        print(" Input harus berupa angka!")

# Contoh penggunaan
hitung_efisiensi_termal()
```

```python


Penjelasannya :

Program di atas merupakan fungsi Python bernama hitung_efisiensi_termal() yang digunakan untuk menghitung efisiensi termal pada turbin gas. Di awal program, pengguna diminta untuk memasukkan dua data, yaitu energi masuk dan energi keluar dalam satuan Joule. Program kemudian menghitung efisiensi dengan rumus (energi keluar / energi masuk) × 100 untuk menghasilkan nilai efisiensi dalam persen. Program juga dilengkapi dengan pemeriksaan kesalahan agar input yang dimasukkan harus berupa angka dan memastikan energi masuk tidak nol atau negatif. Setelah perhitungan, hasil efisiensi ditampilkan bersama analisis otomatis: jika nilainya di atas 70%, sistem dianggap sangat efisien; antara 40–70% dinilai cukup efisien; sedangkan di bawah 40% dianggap kurang efisien. Program ini dibuat interaktif, informatif, dan mudah digunakan untuk memahami konsep efisiensi termal secara praktis.
```

```python
#10. While Loop & Logika
tangki = 0  # mulai dari 0 liter

while tangki <= 100:
    print("Isi tangki:", tangki, "liter")

    if tangki == 80:
        print("WARNING: Hampir penuh!")
    elif tangki == 100:
        print("TANGKI PENUH – Pengisian berhenti.")
        break  # hentikan loop saat tangki penuh

    tangki += 10  # naikkan 10 liter tiap iterasi
```

```python


Penjelasannya :

Program ini mensimulasikan proses pengisian tangki bahan bakar secara bertahap menggunakan perulangan while. Pada awalnya, variabel tangki diatur dengan nilai 0 liter yang merepresentasikan kondisi tangki kosong. Kemudian, program akan terus menjalankan perulangan selama nilai tangki kurang dari atau sama dengan 100. Setiap kali loop berjalan, program menampilkan jumlah bahan bakar yang telah terisi menggunakan perintah print("Isi tangki:", tangki, "liter").

Di dalam loop, terdapat pernyataan if–elif yang digunakan untuk menampilkan pesan peringatan pada kondisi tertentu. Saat nilai tangki mencapai 80 liter, program akan mencetak pesan “WARNING: Hampir penuh!” sebagai tanda bahwa kapasitas tangki hampir maksimal. Lalu, ketika nilai tangki mencapai 100 liter, program menampilkan pesan “TANGKI PENUH – Pengisian berhenti.” dan perulangan dihentikan menggunakan perintah break, yang berfungsi untuk keluar dari loop meskipun kondisi while masih bernilai benar.

Setiap iterasi, nilai variabel tangki bertambah 10 liter melalui perintah tangki += 10. Dengan cara ini, proses simulasi pengisian berlangsung secara bertahap seperti pengisian bahan bakar nyata. Logika ini juga membantu memahami cara kerja loop while yang dikombinasikan dengan kondisi logika (if–elif) untuk mengatur perilaku program sesuai keadaan tertentu. Secara keseluruhan, program ini menunjukkan bagaimana struktur kontrol dalam Python dapat digunakan untuk membuat simulasi yang logis, interaktif, dan mudah dipahami.
```

```python
#CHALLENGE
# PROGRAM PEMANTAUAN ENERGI & PENGISIAN BATERAI PANEL SURYA

# energi hasil konversi dari panel surya
def hitung_energi(input_daya, efisiensi):
    """
    Menghitung energi keluaran (Wh) berdasarkan input daya dan efisiensi sistem.
    """
    energi_output = input_daya * efisiensi
    return energi_output

# Kondisi kapasitas batrai
def status_baterai(kapasitas):
    if kapasitas < 30:
        return "RENDAH "
    elif kapasitas < 80:
        return "NORMAL "
    elif kapasitas < 100:
        return "TINGGI "
    else:
        return "PENUH "


print(" Simulasi Pengisian Batrai Oleh Panel Surya ")

daya_masuk = float(input("Masukkan daya input dari panel surya (Wh): "))
efisiensi = float(input("Masukkan efisiensi sistem (0-1): "))

energi_harian = hitung_energi(daya_masuk, efisiensi)
print(f"\nEnergi yang dihasilkan per siklus: {energi_harian:.2f} Wh\n")

kapasitas_baterai = 0

# simulasi pengisian baterai
while kapasitas_baterai < 100:
    kapasitas_baterai += energi_harian / 10
    if kapasitas_baterai > 100:
        kapasitas_baterai = 100

    print(f"Kapasitas baterai saat ini: {kapasitas_baterai:.2f}% - {status_baterai(kapasitas_baterai)}")

    # notifikasi pengisian batrai
    if kapasitas_baterai >= 90 and kapasitas_baterai < 100:
        print(" WARNING: Baterai hampir penuh!")
    elif kapasitas_baterai == 100:
        print(" Baterai penuh. Pengisian berhenti.")
        break


print("- CATATAN DAYA LISTRIK HARIAN -")

# estimasi energi selama seminggu
for hari in range(1, 8):
    total_energi = energi_harian * hari
    print(f"Hari ke-{hari}: {total_energi:.2f} Wh terkumpul")
```

```python


Penjelasannya :

Program ini dibuat untuk mensimulasikan proses pengisian baterai menggunakan energi dari panel surya. Masalah utama yang diselesaikan adalah bagaimana memantau berapa banyak energi yang dihasilkan panel dan sejauh mana baterai terisi dari waktu ke waktu. Cara kerjanya cukup sederhana. Pertama, fungsi hitung_energi(input_daya, efisiensi) menghitung energi yang dihasilkan panel berdasarkan daya masuk dan efisiensi sistem. Lalu, fungsi status_baterai(kapasitas) digunakan untuk menampilkan kondisi baterai, apakah masih rendah, normal, tinggi, atau sudah penuh. Setelah kita memasukkan daya input dan efisiensi, program menghitung energi per siklus dan mulai mensimulasikan proses pengisian baterai dengan perulangan while kapasitas_baterai < 100: Setiap pengisian akan menambah kapasitas baterai sedikit demi sedikit, sambil menampilkan statusnya. Jika kapasitas mencapai 90%, muncul peringatan “hampir penuh”, dan saat sudah 100%, pengisian otomatis berhenti. Di bagian akhir, program juga menghitung perkiraan total energi yang dikumpulkan selama seminggu menggunakan perulangan for. Bagian kode yang paling penting ada pada: 1.energi_output = input_daya * efisiensi (untuk menghitung energi hasil konversi), 2.while kapasitas_baterai < 100: (untuk simulasi pengisian baterai), 3. status_baterai(kapasitas) (untuk menampilkan kondisi baterai).
```
